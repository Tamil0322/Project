```python
User=input("Enter your username: ")
Password=input("Enter your password: ")
print(f"Welcome  {User}!!,to the Online Shopping System!")
products = {
    "1": {"name": "Laptop", "price": 1200},
    "2": {"name": "Mouse", "price": 25},
    "3": {"name": "Keyboard", "price": 75},
    "4": {"name": "Monitor", "price": 300},
    "5": {"name": "Headphones", "price": 100},
    "6": {"name": "Tablet", "price": 400},
    "7": {"name": "Printer", "price": 150},
    "8": {"name": "External Hard Drive", "price": 80},
    "9": {"name": "Webcam", "price": 50},
    "10": {"name": "Wireless Mouse", "price": 30}
}
cart = []
def show_products():
    for product_id, product_name in products.items():
        print(f"{product_id}: {product_name['name']} - ${product_name['price']}")
def add_to_cart(product_id):
   if product_id in cart:
    cart.append(products[product_id])
    print(f"{product_name['name']} added to cart.")
   else:
    print("Invalid product ID.")
def view_cart():
    if not cart:
        print("Your cart is empty.")
    else:
        total_price = 0
        for item in cart:
            print(f"{item['name']} - ${item['price']}")
            total_price += item['price']
        print(f"Total: ${total_price}")
        print(f"Total amount: ${total_price}")
def checkout():
    if not cart:
        print("Your cart is empty. Please add items before checking out.")
        return
    print("Order placed successfully!")
    cart.clear()
while True:
    print("\nOptions \n1.view\n2.add\n3.cart\n4.checkout\n5.exit\n")
    choice = input("Enter your choice: ")
    if choice == "1":
        show_products()
    elif choice == "2":
        product_id = input("Enter product ID to add: ")
        add_to_cart(product_id)
    elif choice == "3":
        view_cart()
    elif choice == "4":
        checkout()
    elif choice == "5":
        print("Thank you for shopping with us!")
        break
    else:
        print("Invalid choice.")
```
